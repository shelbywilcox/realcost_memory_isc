# boil down to 1 number = viewing, 1 = imagining

```python
view_length = 30

imagine_length = 10
```

```python
view_ad_1 = np.mean(current_time_series[onsets_view[0]: (onsets_view[0]+view_length)], axis = 0)
view_ad_2 = np.mean(current_time_series[onsets_view[1]: (onsets_view[1]+view_length)], axis = 0)
view_ad_3 = np.mean(current_time_series[onsets_view[2]: (onsets_view[2]+view_length)], axis = 0)


plt.plot(view_ad_1)
```

```python
imagine_ad_1 = np.mean(current_time_series[onsets_imagine[0]: (onsets_imagine[0]+imagine_length)], axis = 0)
imagine_ad_2 = np.mean(current_time_series[onsets_imagine[1]: (onsets_imagine[1]+imagine_length)], axis = 0)
imagine_ad_3 = np.mean(current_time_series[onsets_imagine[2]: (onsets_imagine[2]+imagine_length)], axis = 0)
```

```python

plt.plot(view_ad_1)
plt.plot(imagine_ad_1)

np.corrcoef(view_ad_1, imagine_ad_1)
```

```python
vectors = np.vstack((view_ad_1,
                     view_ad_2,
                     view_ad_3,
                     imagine_ad_1,
                     imagine_ad_2,
                     imagine_ad_3) )
plt.imshow(np.corrcoef(vectors), cmap = 'seismic', vmin = -1, vmax = 1);
plt.colorbar()
```
